```python
# Importar as bibliotecas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# Carregar dados
lista = pd.read_csv('unicornio.csv', sep=",", encoding="utf-8", index_col=0)
lista.head()
```

```python
# Inserir nome no Index da tabela
lista.index.name = 'id'
```

```python
lista.head()
```

```python
# Renomear nomes das colunas
lista.rename(columns={
    "Name":"Nome",
    "Valuation":"Valor",
    "company started":"Início da empresa",
    "country name":"País",
    "city":"Cidade",
    "sector":"Setor",
    "investors":"Investidores"
}, inplace = True)

lista.head()
```

```python
#Verificar campos nulos

lista.isnull().sum()
```

```python
#Verificar campos unicos
lista.nunique()
```

```python
# Verificar tipos de dados e campos não nulos
lista.info()
```

```python
# Verificar campos nulos com mapa de calor
plt.figure(figsize=(15,6))
plt.title("Verificando campos nulos")
sns.heatmap(lista.isnull());
```

```python
#Verificar valores únicos na coluna
lista['Setor'].unique()
```

```python
# Valores únicos, mostrando a quantidade de cada
lista['Setor'].value_counts()
```

```python
# Valores únicos, mostrando a quantidade por porcewntagem
lista['Setor'].value_counts(normalize=True)
```

```python
# mostrar dados com plt
plt.figure(figsize=(15,6))
plt.title("Analisando dados")
plt.bar(lista['Setor'].value_counts().index, lista['Setor'].value_counts())
plt.xticks(rotation=45, ha='right');
```

```python
# Arredondar valores com o round e normalize
var = round(lista['País'].value_counts(normalize=True) * 100,1)
var
```

```python
# Gráfico de pizza mostrando top 8 paises unicornios
plt.figure(figsize=(15,6))
plt.title("Análise de paises")
plt.pie(
    var.head(8),
    labels=var.index[0:8],
    shadow=True,
    startangle=90,
    autopct='%1.1f%%'
);
```

```python
# Converter objeto para datatime
lista['Início da empresa']= pd.to_datetime(lista['Início da empresa'], dayfirst=True, errors='coerce')
lista['Início da empresa']
```

```python
# Extrair o mês e ano da data



lista['mês'] = lista['Início da empresa'].dt.strftime('%m')
lista['Ano'] = lista['Início da empresa'].dt.strftime('%Y')


lista.head()
```

```python
# Tábela análitica de dados
analise = lista.groupby(by=['País','Ano','mês']).count()
```
